**Project Proposal:**

A triathlon is a multisport race with a swimming, biking, and running leg. There are different distance options ranging from a sprint triathlon (which takes about two hours) to an iron man (a true endurance test which takes about 13 hours). This year, I trained for and ran my first ever sprint triathlon. As a college student, I am constantly juggling my academic, professional, and athletic goals. 

I am hoping to discover ways to make triathlon training more efficient. If an individual can input two of their times in individual legs (i.e swimming and running) - it would input their sample bike time. Using this, athletes can focus their training on a particular discipline that is comparatively weaker.

Additionally, triathlons are an expensive hobby. I, for one, did not have access to a bike for the entirity of my training. This model would give athletes an idea of their fitness levels if they did not have access to a bike or a pool.

A second aspect of my predictions is going to involve a categorical variable I created around age. There doesn't appear to be a trend between age, as a category, and overall finish time so it is interesting if the model including age is better than the one that doesn't. Finaly, I will be using the model that uses swimming and running as the predictors for the model comparison.

**Data Description:**

The data sets includes basic information about each participant and their overall, swim, bike, and run times - with finishing rankes included from their race year. It also includes information about transition times (time spent between the different sports that does count towards a participant's overall time).

The data was collected through a tracker (worn on each participant's left ankle). These trackers don't always work in measuring the individual aspects of the race. When I did it this last year, there were some weird instructions that were easy to mess up in terms of steps during the transition times. I had to drop a lot of observations that had hour-long swimming times or transition times; this means that the tracker did not catch them switching disciplines from swimming to biking, etc.

**Main Variable Description:**

The data was originally formatted in hours:minutes:seconds. One sample finish time could be 1:12:45, meaning one hour, twelve minutes, and forty five seconds. I will go into this more in the next part of the executive summary but I needed these variables to be a numeric type for analyzation and visualization.

The variables I've taken an extended look at it (in both the visualization and the machine learning portions) include an athlete's overall finish time, swimming time, biking time, running time, transition times, and age. Age has been stored as both a numeric and a categorical variable.

**Decription of Data Cleaning:**

I had to do a lot of cleaning with these data sets. There were a lot of issues with the formatting of the data that needed to be addressed. I don't need to go into each minute detail, but the most important part was changing each of the overall time, swimming time, biking time, running time, and transition times into numeric variables. However, some participants had swimming times that were incredibly long, due to issues with the ankle tracker, so they needed to be isolated and removed. The same with some of the Transition 1 times and the biking times.

Some other issues included the formatting for some of the time variables. Specifically, the bike times were formatted as such: 32:12:00. It is obvious that the bikers did not take thirty two hours and twelve minutes to complete their ride and that the data is meant to convey thirty two minutes and twelve seconds. Every time variable had its own quirks that needed to be addressed and standardized.

**Description of Machine Learning:**

I began the machine learning portion of the project looking to answer a question I was curious about. It doesn't constitute the bulk of my project proposal but I wanted to see how well a simple linear regression with two of the legs would predict an athlete's finish in the third. The more exciting part about that was seeing where my observation gets predicted and where I need to improve. In the figure below, you see the predicted time for my biking in comparison to where I actually was. If I run and train for a future race, this model shows that I am biking at a much slower pace in comparison to my swimming and biking times.

visualization.png

Into the bulk of the machine learning, I wanted to use swimming and running times to predict overall finish times. I chose running and swimming because they are my two favorite sports and what I like to train in. Additionally, it is easiest to lower your biking time (as it is the longest discipline).

I created three models and wanted to see which had the lowest RMSE. The three are listed below:

Model 1: 2 Quantitative Predictors w/ Linear Regression

Model 2: 2 Quantitative Predictors w/ kNN (k=10)

Model 3: 2 Quantitative Predictors and a Cat Predictor w/ kNN (k=3)

The optimal k-values were discovered using Hyperparameter Tuning. The cross validation estimates of test error for the three models are listed below.

Model 1: 6.015

Model 2: 6.53

Model 3: 7.11

Using the cross validation estimates of test error for each of the three models, we see that Model 1 has the lowest RMSE. The simplest model, solely utilizing Linear Regression, is off on predicting the finish time by an average of 6.015 minutes, compared to 6.53 and 7.11 minutes done by the other two models. The data we have is very linear; the faster you do the running, swimming, or biking portion, the faster your overall time. The most interesting observation from the machine learning aspect of this project is that age is just a number when it comes to triathletes. The model that took age into account performed significantly worse than both models that ignored that variable.
