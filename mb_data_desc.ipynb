**Morro Bay Triathlon Data Description**

Visualizing the webscraped and cleaned data from the data collection file.

```python
import altair as alt
import pandas as pd
```

```python
df_tri = pd.read_csv("morro_triathlon.csv")
```

```python
alt.Chart(df_tri).mark_point().encode(
    x="Swim Time",
    y="Time",
    color="AgeCat",
    tooltip=["Last Name", "Swim Time", "Time", "AgeCat"]
).properties(
    title="Swim Time vs Overall Time"
).interactive()
```

```python
alt.Chart(df_tri).mark_point().encode(
    x="Bike Time",
    y="Time",
    color="AgeCat",
    tooltip=["Last Name", "Bike Time", "Time", "AgeCat"]
).properties(
    title="Bike Time vs Overall Time"
).interactive()
```

```python
alt.Chart(df_tri).mark_point().encode(
    x="Run Time",
    y="Time",
    color="AgeCat",
    tooltip=["Last Name", "Run Time", "Time", "AgeCat"]
).properties(
    title="Run Time vs Overall Time"
).interactive()
```

```python
alt.Chart(df_tri).mark_bar().encode(
    x="AgeCat",
    y="Time"
).properties(
    title="Overall Time Across Age Categories"
)
```

```python
age_order = ["10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79"]

df_tri["AgeCat"].value_counts().loc[age_order].plot.bar(title="Triathlete Age Distribution")
```

```python
alt.Chart(df_tri).mark_point().encode(
    x="Age",
    y="Time",
    tooltip=["Last Name", "Time", "Age"]
).properties(
    title="Morro Bay Triathlon Finish Times Compared to Age"
).interactive()
```

```python
alt.Chart(df_tri).mark_point().encode(
    x="Run Time",
    y="Time",
    facet="AgeCat",
    tooltip=["Last Name", "Time", "Run Time", "Age"]
).properties(
    title="Morro Bay Triathlon Finish Times Compared to Age"
).interactive()
```
